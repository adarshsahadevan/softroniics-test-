```python
def rot(nums, k):
    n = len(nums)
    rotd = [0] * n
    for i in range(k):
      rotd[i] = nums[n - k + i]
    for i in range(n - k):
      rotd[k + i] = nums[i]
    return rotd


nums1 = [1, 2, 3, 4, 5, 6, 7]
k1 = 3
print(rot(nums1, k1))

nums2 = [-1, -100, 3, 99]
k2 = 2
print(rot(nums2, k2))
```
